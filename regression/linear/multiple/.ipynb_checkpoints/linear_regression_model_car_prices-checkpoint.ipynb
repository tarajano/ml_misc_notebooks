### Running normality tests on data
After visualizing data for normality, we want to conduct some actual tests

**The Shapiro-Wilk test** evaluates a data sample and quantifies how likely it is that the data was drawn from a Gaussian distribution. 

The null-hypothesis of this test is that the population is normally distributed. Thus, on the one hand, if the p-value is less than the chosen alpha level, then the null hypothesis is rejected and there is evidence that the data tested are not normally distributed. On the other hand, if the p-value is greater than the chosen alpha level, then the null hypothesis that the data came from a normally distributed population can not be rejected (e.g., for an alpha level of 0.05, a data set with a p-value of less than 0.05 rejects the null hypothesis that the data are from a normally distributed population) https://en.wikipedia.org/wiki/ShapiroWilk_test

In practice, the this test is believed to be a reliable test of normality, although there is some suggestion that the test may be suitable for smaller samples of data, e.g. thousands of observations or fewer.

(More tests for normality can be foud at: https://machinelearningmastery.com/a-gentle-introduction-to-normality-tests-in-python/)

```python
def shapiro_wilks_test(data):
    '''
        Returns a list: 
        statistic value and p-value
    '''
    from scipy.stats import shapiro
    stat, p = shapiro(data)
    return [stat, p]

def test_H0(p_value):
    alpha = 0.05
    if p_value > alpha:
        print('Sample looks Gaussian (fail to reject H0 at alpha={:.3f})'.format(alpha))
    else:
        print('Sample does not look Gaussian (reject H0 at alpha={:.3f})'.format(alpha))
    print('')

sw_stat, p_value = shapiro_wilks_test(dtf.mileage)
print('Mileage sw_stat={:.3f}, p_value={:.3f}'.format(sw_stat, p_value))
test_H0(p_value)

sw_stat, p_value = shapiro_wilks_test(dtf.age_yrs)
print('Age sw_stat={:.3f}, p_value={:.3f}'.format(sw_stat, p_value))
test_H0(p_value)

sw_stat, p_value = shapiro_wilks_test(dtf.sell_price)
print('Sell price sw_stat={:.3f}, p_value={:.3f}'.format(sw_stat, p_value))
test_H0(p_value)
```

### Identifying collinearity between independent variables

Linear Regression assumes lack of collinearity between dependent variables. Collinearity cant negatively affect the performance of the model.

```python
plt.scatter(dtf.mileage, dtf.age_yrs, color='m')
plt.title('mileage vs age_yrs')
plt.xlabel('mileage')
plt.ylabel('age_yrs')
plt.show()

mil_yrs_lrm = LinearRegression()
mil_yrs_lrm.fit(dtf[['mileage']].values, dtf.age_yrs)
r2 = mil_yrs_lrm.score(dtf[['mileage']].values, dtf.age_yrs)
print('mileage vs age_yrs R^2', r2)

""" There is a strong correlation (collinearity also) between mileage and age_yrs. We must avoid including both variables in the model. """ 
```

### Quantifying collinearity. The Variance Inflation Factor (VIF)

```python
"""
Implementation:
    https://etav.github.io/python/vif_factor_python.html (code below)

Interpretation: 
    https://www.displayr.com/variance-inflation-factors-vifs/
    A value of 1 means that the predictor is not correlated with other variables.
    The higher the value, the greater the correlation of the variable with other variables.
    Values of more than 4 or 5 are sometimes regarded as being moderate to high, with values
    of 10 or more being regarded as very high. These numbers are just rules of thumb.

Workarounds: 
    https://www.statisticssolutions.com/assumptions-of-multiple-linear-regression/
    If multicollinearity is found in the data, one possible solution is to center the data.
    To center the data, subtract the mean score from each observation for each independent variable.
    However, the simplest solution is to identify the variables causing multicollinearity
    issues (i.e., through correlations or VIF values) and removing those variables from the regression.
"""

""" Computing VIFs """
vif_dtf = dtf[['mileage', 'age_yrs', 'sell_price']].copy()
vif_res_dtf = pd.DataFrame()
vif_res_dtf["features"] = vif_dtf.columns
vif_res_dtf['vif_factors'] = [variance_inflation_factor(vif_dtf.values, i) for i in range(vif_dtf.shape[1])]
print(vif_res_dtf)

""" Both mileage and age_yrs have very large VIFs, we must avoid including them at the same time in a the model """
```

### Simple Linear Regression

```python
""" sell_price vs mileage """
mil_lrm = LinearRegression()
mil_lrm.fit(dtf[['mileage']].values, dtf.sell_price)
r2 = mil_lrm.score(dtf[['mileage']].values, dtf.sell_price)
sell_price_mil_pred_err = dtf.sell_price.values - mil_lrm.predict(dtf[['mileage']].values)
print('sell_price vs mileage R^2', r2)

""" sell_price vs age """
age_lrm = LinearRegression()
age_lrm.fit(dtf[['age_yrs']].values, dtf.sell_price)
r2 = age_lrm.score(dtf[['age_yrs']].values, dtf.sell_price)
sell_price_age_pred_err = dtf.sell_price.values - age_lrm.predict(dtf[['age_yrs']].values)
print('sell_price vs age R^2', r2)

''' Not bad R2 results for such a few data points! '''
```

### Testing normality of residuals

**Residuals** or errors, are the differences between the predicted and real values of the independent variable.

Simple Linear Regression assumes residuals are normaly distributed. 

```python
# Histograms
plt.hist(sell_price_mil_pred_err, color='g')
plt.title('Mileage - Residuals Hist')
plt.xlabel('Residuals')
plt.show()

plt.hist(sell_price_age_pred_err, color='g')
plt.title('Age - Residuals Hist')
plt.xlabel('Residuals')
plt.show()

# QQ-plot
qqplot(sell_price_mil_pred_err, line='s', color='y')
plt.title('Mileage - Residuals QQ-plot')
plt.show()

qqplot(sell_price_age_pred_err, line='s', color='y')
plt.title('Age - Residuals QQ-plot')
plt.show()

# PP-plots
stats.probplot(sell_price_mil_pred_err, dist="norm", plot=pylab)
pylab.title('Mileage - Residuals PP plot')
pylab.show()

stats.probplot(sell_price_age_pred_err, dist="norm", plot=pylab)
pylab.title('Age - Residuals PP plot')
pylab.show()

''' Seems that residuals may fit to normality. Let's verify it by running a normality test. '''
```

```python
print('Shapiro-Wilks normality results for residuals\n')

sw_stat, p_value = shapiro_wilks_test(sell_price_mil_pred_err)
print('  Mileage sw_stat={:.3f}, p_value={:.3f}'.format(sw_stat, p_value))
test_H0(p_value)

sw_stat, p_value = shapiro_wilks_test(sell_price_age_pred_err)
print('  Age sw_stat={:.3f}, p_value={:.3f}'.format(sw_stat, p_value))
test_H0(p_value)

''' That's great! Our residuals comply with the assumption of normal distribution '''
```

### Testing the assumption of homoscedastacity (lack of heteroscedastacity ) of residuals

**Residuals** or errors, are the differences between the predicted and real values of the independent variable.

**Homoscedastacity** means that residuals exhibit constant variance. This is a key assumption of linear regression. 

**Plots** A residuals plot of homoscedastatic points should look shot out of a shotgun, lacking any obvious pattern, with points equally distributed above and below zero on the X axis, and to the left and right of zero on the Y axis. On the other hand, its a sign you have heteroscedasticity if your residual plot has fan-like shape, or show some sort of growth.

```python
# You can plot Residuals vs. X OR vs. Predicted (^y), but you should never plot Residuals vs y.

''' Sell Price Residuals - Mileage Model '''
plt.scatter(mil_lrm.predict(dtf[['mileage']].values),
            sell_price_mil_pred_err,
            color='r')
plt.title('Sell Price Residuals - Mileage Model')
plt.xlabel('Predicted')
plt.ylabel('Residuals')
plt.show()

''' Sell Price Residuals - Age Model '''
plt.scatter(age_lrm.predict(dtf[['age_yrs']].values),
            sell_price_age_pred_err,
            color='g')
plt.title('Sell Price Residuals - Age Model')
plt.xlabel('Predicted')
plt.ylabel('Residuals')
plt.show()

'''Residuals look fairly random for both models. That is good news for the homoscedastacity assumption. However we need to test that thouroghly with a statistic analysis. '''
```

### Let's test and quantify homoscedastacity using BreuschPagan test

**BreuschPagan test** is used to test for heteroskedasticity in a linear regression model. It tests whether the variance of the errors from a regression is dependent on the values of the independent variables. In that case, heteroskedasticity is present. *If the test statistic has a p-value below an appropriate threshold (e.g. p<0.05) then the null hypothesis of homoskedasticity is rejected and heteroskedasticity assumed* (https://en.wikipedia.org/wiki/BreuschPagan_test)

```python
# http://www.statsmodels.org/stable/diagnostic.html    
from statsmodels.stats.diagnostic import het_breuschpagan
alpha = .05

print('Heteroscedasticity Tests')

# Heteroscedasticity test for - Mileage Model 
sell_price_mil_pred = mil_lrm.predict(dtf[['mileage']]).reshape(-1,1)
_, _, fvalue, f_pvalue = het_breuschpagan(sell_price_mil_pred_err, sell_price_mil_pred)
print('Mileage Model\tfvalue:{:.3f} f_pvalue:{:.3f}'.format(fvalue, f_pvalue))

# Heteroscedasticity test for - Age Model
sell_price_age_pred = mil_lrm.predict(dtf[['age_yrs']]).reshape(-1,1)
_, _, fvalue, f_pvalue = het_breuschpagan(sell_price_age_pred_err, sell_price_age_pred)
print('Age Model\tfvalue:{:.3f} f_pvalue:{:.3f}'.format(fvalue, f_pvalue))

"""Both models fail the assumption of homoscedastacity at alpha={:.2f}. \
However we could assume homoscedastacity for the Mileage model at alpha=0.01.""".format(alpha)
```

### Applying a polynomial regression

```python
def poly_reg(X, y):
    X_train = X
    y_train = y

    ## Fits Linear Regression to the Training set
    from sklearn.linear_model import LinearRegression
    slm_classifier = LinearRegression()
    slm_classifier.fit(X_train, y_train)

    ## Transforms features into polynomial
    from sklearn.preprocessing import PolynomialFeatures
    poly_reg = PolynomialFeatures(degree=4)
    # Adds additional polynomial terms (transaformed features) to the features matrix
    X_poly = poly_reg.fit_transform(X_train)

    ## Fits the polynomial features to a LinearRegression 
    #  Needs to use a different LinearRegression obj
    slm_classifier2 = LinearRegression()
    slm_classifier2.fit(X_poly, y_train)
    print('R^2:', slm_classifier2.score(X_poly, y_train))
    
poly_reg(X, dtf.sell_price)
poly_reg(Xyrs, dtf.sell_price)
poly_reg(Xmil, dtf.sell_price)
```
